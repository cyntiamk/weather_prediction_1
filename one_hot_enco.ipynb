```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('kyoto_weather.csv')

weather = file.set_index('Date')

weather.head()
```

```python
# group by description to find out all categories labels necessary to hot-encode
ohe = weather.groupby(['Description']).mean()
ohe
```

```python
# one-hot-enconding classes
# 0. Clear
# 1. Clouds
# 2. Fog
# 3. Haze
# 4. Mist
# 5. Rain
# 6. Snow
# 7. Thunderstorm
```

```python
# Step 0: Reformat data
data = weather.values
X = data[:, 0:10]
y = data[:, 10]
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```
