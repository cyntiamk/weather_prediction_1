```python
import pandas as pd
import numpy as np
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
may = 'https://flights.msn.com/en-us/flight-search#/result?originplace=LAX&destinationplace=KIX&outbounddate=2019-05-01&inbounddate=&cabinclass=Economy&adults=1&children=0&infants=0'
!which chromedriver
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit(may)
```

```python
may = []
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
grab_div = soup.find('ul', class_='chart-content')
list_item = grab_div.find_all('li')
for item in list_item:
    may.append(item['title'])
pd.DataFrame(may)
```
